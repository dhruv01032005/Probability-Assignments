```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
players=pd.read_csv("/content/drive/MyDrive/atp_players.csv",sep =",")
data1=pd.read_excel("/content/drive/MyDrive/atp_2022.xlsx")
data2=pd.read_excel("/content/drive/MyDrive/atp_2023.xlsx")
data3=pd.read_excel("/content/drive/MyDrive/atp_2024.xlsx")
```

## Question 1: Find the probability and CDF of all the birthyear in the apt_players database.

```python
birthyear=players["birth_year"]
l1=[]
l2=[]
c=0
for k in birthyear:
  if np.isnan(k):
    continue
  elif k not in l1:
    l1.append(k)
    l2.append(1)
    c=c+1
  else:
    ind=l1.index(k)
    l2[ind]=l2[ind]+1
    c=c+1
for k in range(len(l2)):
  l2[k]=l2[k]/c

for k in range(1,len(l2)):
  for j in range(len(l2)-k):
    if l1[j]>l1[j+1]:
      l1[j],l1[j+1]=l1[j+1],l1[j]
      l2[j],l2[j+1]=l2[j+1],l2[j]
d=0
cdf=[]
print("\n\nPROBABILITY")
for k in range(len(l1)):
  print(l1[k]," ",l2[k])
  d=d+l2[k]
  cdf.append(d)
print("\n\n\nCDF")
for k in range(len(cdf)):
  print(l1[k]," ",cdf[k])
```

## Question2: Plot the PMF and CDF for the above data

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.bar(l1,l2,color="blue",alpha=0.5)
ax1.set_xlabel("year")
ax1.set_ylabel("probability")
ax1.set_title("Probability Table")

ax2.bar(l1,cdf,color="red",alpha=0.5)
ax2.set_xlabel("year")
ax2.set_ylabel("cdf")
ax2.set_title("Probability Table")
```

## Question3: Find the Joint PMF and Joint CDF for the W1 and W2 in database atp_2022.

```python
w1=data1["W1"]
w2=data1["W2"]

value_1=[1,2,3,4,5,6,7]
value_2=[1,2,3,4,5,6,7]

jp=np.zeros([7,7])
c=0
for k in range(len(w1)):
  if np.isnan(w1[k]) or np.isnan(w2[k]):
    continue
  else:
    jp[int(w2[k])-1][int(w1[k])-1]=jp[int(w2[k])-1][int(w1[k])-1]+1
    c=c+1
for k in range(7):
  for j in range(7):
    jp[k][j]=jp[k][j]/c
print("JOINT PMF")
print(jp)

jc=np.zeros([7,7])
for k in range(7):
  for j in range(7):
    c=0
    for i in range(k+1):
      for l in range(j+1):
        c=c+jp[i][l]
    jc[k][j]=c
print("JOINT CDF")
print(jc)
```

## Question4: Plot the JOINT PMF and JOINT CDF for the above data

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')
_xx, _yy = np.meshgrid(list(range(1,len(value_1)+1)), list(range(1,len(value_2)+1)))
_x, _y = _xx.ravel(), _yy.ravel()
j_p=np.array(jp)
j_p=j_p.ravel()
j_c=np.array(jc)
j_c=j_c.ravel()
bottom = np.zeros(len(value_1)*len(value_2))
width = depth = 1

ax1.bar3d(_x, _y, bottom, width, depth, j_p, shade=True)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Probability")
ax1.set_title('JOINT PMF')

ax2.bar3d(_x, _y, bottom, width, depth, j_c, shade=True)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Cummulative Function")
ax2.set_title('JOINT CDF')
```

## Question7: Find the correlation coefficient of weight_lbs and weight_kg in the database atp_players.

```python
lbs=players["weight_lbs"]
kg=players["weight_kg"]

lbs_mean=0
c=0
for k in lbs:
  if np.isnan(k):
    continue
  else:
    lbs_mean=lbs_mean+k
    c=c+1
lbs_mean=lbs_mean/c

kg_mean=0
c=0
for k in kg:
  if np.isnan(k):
    continue
  else:
    kg_mean=kg_mean+k
    c=c+1

kg_mean=kg_mean/c

e=0
for k in range(len(lbs)):
  if np.isnan(lbs[k]) or np.isnan(kg[k]):
    continue
  else:
    e=e+lbs[k]*kg[k]
    c=c+1

e=e/c

lbs_variance=0
c=0
for k in lbs:
  if np.isnan(k):
    continue
  else:
    lbs_variance=lbs_variance+(k-lbs_mean)**2
    c=c+1
lbs_variance=lbs_variance/c

kg_variance=0
c=0
for k in kg:
  if np.isnan(k):
    continue
  else:
    kg_variance=kg_variance+(k-kg_mean)**2
    c=c+1
kg_variance=kg_variance/c

cv=e-lbs_mean*kg_mean
cc=cv/(((lbs_variance)**(1/2))*((kg_variance)**(1/2)))
print("Mean of lbs",lbs_mean)
print("Mean of kg",kg_mean)
print("Expectation of lbs and kg",e)
print("Covariance",cv)
print("Correlation",cc)
```

## Question8: Find the normalized p where p is the database containing WRank, LRank, WPts, LPts of the people in first round from the database atp_2024.

```python
data_1=data3[data3['Round']=='1st Round']
data=data_1[['WRank','LRank','WPts','LPts']]
p=data.values

mean=[]
for k in range(len(p[0])):
  c=0
  d=0
  for j in range(len(p)):
    if np.isnan(p[j][k]):
      p[j][k]=0
      d=d+1
    else:
      c=c+p[j][k]
      d=d+1
  mean.append(c)

for k in range(len(mean)):
  mean[k]=mean[k]/d

variance=[]
for k in range(len(p[0])):
  c=0
  d=0
  for j in range(len(p)):
    if np.isnan(p[j][k]):
      p[j][k]=0
      d=d+1
    else:
      c=c+((p[j][k]-mean[k])**2)
      d=d+1
  variance.append(c)

for k in range(len(variance)):
  variance[k]=variance[k]/d

normal=[]
for k in range(len(p)):
  temp=[]
  for j in range(len(p[0])):
    temp.append((p[k][j]-mean[j])/((variance[j])**(1/2)))
  normal.append(temp)
normal=np.array(normal)
print("Normalized point")
print(normal)
```

## Question9: Reduce the dimentionality to 3 using PCA.


```python
mean_2=[]
for k in range(len(normal[0])):
  c=0
  for j in range(len(normal)):
    c=c+normal[j][k]/len(normal)
  mean_2.append(c)

variance_2=[]
for k in range(len(normal[0])):
  c=0
  for j in range(len(normal)):
    c=c+((normal[j][k]-mean_2[k])**2)/len(normal)
  variance_2.append(c)

cov=[]
for k in range(len(normal[0])):
  temp=[]
  for j in range(len(normal[0])):
    c=0
    for i in range(len(normal)):
      c=c+(normal[i][k]-mean_2[k])*(normal[i][j]-mean_2[j])/len(normal)
    temp.append(c)
  cov.append(temp)
cov=np.array(cov)

eigen=np.linalg.eig(cov)
eigen_values=eigen.eigenvalues
eigen_vectors=eigen.eigenvectors

d_3=eigen_vectors[:,:3]

mul_3=[]
for k in range(len(normal)):
  temp=[]
  for j in range(len(d_3[0])):
    c=0
    for i in range(len(normal[0])):
      c=c+normal[k][i]*d_3[i][j]
    temp.append(c)
  mul_3.append(temp)
mul_3=np.array(mul_3)
print("3D dimension reduction")
print(mul_3)
```

## Question10: Plot the data reduced to 3D.


```python
x=[]
y=[]
z=[]
for k in range(len(mul_3)):
  x.append(mul_3[k][0])
  y.append(mul_3[k][1])
  z.append(mul_3[k][2])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=x,ys=y,zs=z)
ax.set_title("3D dimentional reduction")
ax.set_xlabel("Principle componenet 1")
ax.set_ylabel("Principle componenet 2")
ax.set_zlabel("Principle componenet 3")
ax.legend()
plt.show()
```
