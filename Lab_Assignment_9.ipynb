```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset(name='iris')
data
```

```python
# finding p
data_2=data[['sepal_length','sepal_width','petal_length','petal_width']]
p=data_2.values
print("Sample data:")
print(p)
```

```python
l_mean=[]
for k in range(len(p[0])):
  c=0
  for j in range(len(p)):
    c=c+p[j][k]/len(p)
  l_mean.append(c)
```

```python
l_variance=[]
for k in range(len(p[0])):
  c=0
  for j in range(len(p)):
    c=c+((p[j][k]-l_mean[k])**2)/len(p)
  l_variance.append(c)
```

```python
# Step 1: Normalization
normal=[]
for k in range(len(p)):
  temp=[]
  for j in range(len(p[0])):
    temp.append((p[k][j]-l_mean[j])/((l_variance[j])**(1/2)))
  normal.append(temp)
normal=np.array(normal)
print("Normalized point")
print(normal)
```

```python
l_mean_2=[]
for k in range(len(normal[0])):
  c=0
  for j in range(len(normal)):
    c=c+normal[j][k]/len(normal)
  l_mean_2.append(c)

l_variance_2=[]
for k in range(len(normal[0])):
  c=0
  for j in range(len(normal)):
    c=c+((normal[j][k]-l_mean_2[k])**2)/len(normal)
  l_variance_2.append(c)
```

```python
# Step-2: Compute the Covariance matrix
cov=[]
for k in range(len(normal[0])):
  temp=[]
  for j in range(len(normal[0])):
    c=0
    for i in range(len(normal)):
      c=c+(normal[i][k]-l_mean_2[k])*(normal[i][j]-l_mean_2[j])/len(normal)
    temp.append(c)
  cov.append(temp)
cov=np.array(cov)
print(cov)
```

```python
# step-3: Compute the eigen values and the eigen vectors

eigen=np.linalg.eig(cov)
eigen_values=eigen.eigenvalues
eigen_vectors=eigen.eigenvectors
print(eigen_values)
print(eigen_vectors)
```

```python
print("PCA:")
for k in range(len(eigen_values)):
  print(f"PCA {k+1}:",eigen_vectors[:,k])
```

```python
# step-4: dimensionality reduction to 2D
d_2=eigen_vectors[:,:2]

mul_2=[]
for k in range(len(normal)):
  temp=[]
  for j in range(len(d_2[0])):
    c=0
    for i in range(len(normal[0])):
      c=c+normal[k][i]*d_2[i][j]
    temp.append(c)
  mul_2.append(temp)
mul_2=np.array(mul_2)
print("2D dimension reduction")
print(mul_2)
```

```python
# step-4: dimensionality reduction to 3D
d_3=eigen_vectors[:,:3]

mul_3=[]
for k in range(len(normal)):
  temp=[]
  for j in range(len(d_3[0])):
    c=0
    for i in range(len(normal[0])):
      c=c+normal[k][i]*d_3[i][j]
    temp.append(c)
  mul_3.append(temp)
mul_3=np.array(mul_3)
print("3D dimension reduction")
print(mul_3)
```

```python
color=['blue','red','green']
label=['setosa','versicolor','virginica']
x=[]
y=[]
for k in range(len(mul_2)):
  x.append(mul_2[k][0])
  y.append(mul_2[k][1])
for k in range(3):
  plt.scatter(x=x[50*k:50*(k+1)],y=y[50*k:50*(k+1)],color=color[k],label=label[k])
plt.title("2D dimentional reduction")
plt.xlabel("Principle componenet 1")
plt.ylabel("Principle componenet 2")
plt.legend()
plt.show()
```

```python
color=['blue','red','green']
label=['setosa','versicolor','virginica']
x=[]
y=[]
z=[]
for k in range(len(mul_3)):
  x.append(mul_3[k][0])
  y.append(mul_3[k][1])
  z.append(mul_3[k][2])
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for k in range(3):
  ax.scatter(xs=x[50*k:50*(k+1)],ys=y[50*k:50*(k+1)],zs=z[50*k:50*(k+1)],color=color[k],label=label[k])
ax.set_title("3D dimentional reduction")
ax.set_xlabel("Principle componenet 1")
ax.set_ylabel("Principle componenet 2")
ax.set_zlabel("Principle componenet 3")
ax.legend()
plt.show()
```
