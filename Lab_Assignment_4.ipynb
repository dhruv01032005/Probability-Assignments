```python
import matplotlib.pyplot as plt
def uniform_continuous_random_variable():
  a=int(input("Starting of the elements::"))
  b=int(input("Ending of the elements(>a)::"))
  l=[a]
  for k in range(a,b+1):
    l.append(k)
  l.append(b)
  pro=1/(b-a)
  p=[0]
  for k in range(a,b+1):
    p.append(pro)
  p.append(0)
  plt.plot(l,p)
  plt.fill_between(l,p,color='green',alpha=0.2)
  plt.title("Probability Density Function of Uniform Continuous Random Variable")
  plt.xlabel("x")
  plt.ylabel("Probabilty")
  plt.show()
uniform_continuous_random_variable()
```

```python
import matplotlib.pyplot as plt
import math
import numpy as np
def laplacian_continuous_random_variable():
  a=int(input("Starting of the elements::"))
  b=int(input("Ending of the elements(>a)::"))
  mu = float(input("Enter mean:: "))
  b0 = float(input("Enter scale parameter:: "))
  l=[]
  for k in range(a*1000,(b*1000)+1):
    l.append(k/1000)
  fx=[]
  for k in l:
    fx.append(np.where(k < mu, 0.5 * np.exp((k - mu) / b0), 1 - 0.5 * np.exp(-(k - mu) / b0)))
  plt.plot(l,fx)
  plt.fill_between(l,fx,color='green',alpha=0.2)
  plt.title("Cummulative distribution function of Laplacian Continuous Random Variable")
  plt.xlabel("x")
  plt.ylabel("Cummulative Function")
  plt.show()
laplacian_continuous_random_variable()
```

```python
import matplotlib.pyplot as plt
def bernoulli():
  l=[]
  n=int(input("Number of elements::"))
  k=0
  while k<n:
    a=int(input("Element::"))
    if a==0 or a==1:
      l.append(a)
      k=k+1
    else:
      print("Enter 0 or 1")
  l2=[]
  l3=[-1,0,0,1,1,2]
  p=[]
  for k in range(n):
    if l[k] not in l2:
      l2.append(l[k])
      p.append(1/n)
    else:
      ind=l2.index(l[k])
      p[ind]=p[ind]+1/n
  p2=[0,0,p[0],p[0],1,1]
  plt.plot(l3,p2,color='green')
  plt.fill_between(l3,p2,color='green',alpha=0.2)
  plt.title("Cummulative distribution function of Bernoulli Random Variable")
  plt.xlabel("x")
  plt.ylabel("Cummulative Function")
  plt.show()

bernoulli()
```

```python
import matplotlib.pyplot as plt
def factorial(n):
  fac=1
  for k in range(1,n+1):
    fac=fac*k
  return fac
def binomial():
  n=int(input("Number of elements::"))
  l=[]
  for k in range(n+1):
    l.append(k)
  p=float(input("Probability::"))
  bino=[]
  c=0
  for k in range(n+1):
    c=c+(factorial(n)/(factorial(n-l[k])*factorial(l[k]))*((p)**(l[k]))*((1-p)**(n-l[k])))
    bino.append(c)
  plt.step(l,bino,color='red',alpha=0.6)
  plt.title("Cummulative distribution function of Binomial Random Variable")
  plt.xlabel("x")
  plt.ylabel("Cummulative Function")
  plt.show()

binomial()
```

```python
import matplotlib.pyplot as plt
def factorial(n):
  fac=1
  for k in range(1,n+1):
    fac=fac*k
  return fac
def poisson():
  p=float(input("Enter the probability(p-->0)::"))
  m=int(input("Enter the number of elements(m-->infinity)"))
  lamda=p*m
  l=[]
  for k in range(m):
    l.append(k)
  poisson=[]
  c=0
  for k in range(m):
    c=c+((2.718)**(-1*lamda)*((lamda)**k/factorial(k)))
    poisson.append(c)
  plt.step(l,poisson,color='orange',alpha=0.6)
  plt.title("Cummulative distribution function of Poisson Random Variable")
  plt.xlabel("x")
  plt.ylabel("Cummulative Function")
  plt.show()

poisson()
```

```python
import matplotlib.pyplot as plt
def geometric():
  n=int(input("Enter the number of elements::"))
  l=[]
  for k in range(1,n+1):
    l.append(k)
  p=float(input("Enter the probability::"))
  geometric=[]
  c=0
  for k in range(1,n+1):
    c=c+(((1-p)**(k-1))*p)
    geometric.append(c)
  plt.step(l,geometric,color='pink')
  plt.title("Cummulative distribution function of Geometric Random Variable")
  plt.xlabel("x")
  plt.ylabel("Cummulative Function")
  plt.show()

geometric()
```
