```python
import matplotlib.pyplot as plt
def uniform_discrete_function():
  a=int(input("Starting of the elemnets::"))
  b=int(input("Ending of the elemnets::"))
  l=[]
  for k in range(a,b+1):
    l.append(k)
  pro=1/(b-a+1)
  p=[]
  for k in range(a,b+1):
    p.append(pro)
  plt.bar(l,p,color='blue',alpha=0.6)
  plt.title("Probability Mass Function of Uniform Discrete Function")
  plt.xlabel("Data Points")
  plt.ylabel("Probabilty")
  plt.show()

uniform_discrete_function()
```

```python
import matplotlib.pyplot as plt
def bernoulli():
  l=[]
  n=int(input("Number of elements::"))
  k=0
  while k<n:
    a=int(input("Element::"))
    if a==0 or a==1:
      l.append(a)
      k=k+1
    else:
      print("Enter 0 or 1")
  l2=[]
  p=[]
  for k in range(n):
    if l[k] not in l2:
      l2.append(l[k])
      p.append(1/n)
    else:
      ind=l2.index(l[k])
      p[ind]=p[ind]+1/n
  plt.bar(l2,p,color='green',width=0.25,alpha=0.6)
  plt.title("Probability Mass Function of Bernoulli Random Variable")
  plt.xlabel("Sample Space")
  plt.ylabel("Probabilty")
  plt.show()

bernoulli()
```

```python
import matplotlib.pyplot as plt
def factorial(n):
  fac=1
  for k in range(1,n+1):
    fac=fac*k
  return fac
def binomial():
  n=int(input("Number of elements::"))
  l=[]
  for k in range(n+1):
    l.append(k)
  p=float(input("Probability::"))
  bino=[]
  for k in range(n+1):
    bino.append((factorial(n)/(factorial(n-l[k])*factorial(l[k]))*((p)**(l[k]))*((1-p)**(n-l[k]))))
  print(sum(bino))

  plt.bar(l,bino,color='red',alpha=0.6,width=0.5)
  plt.title("Probability Mass Function of Binomial Random Variable")
  plt.xlabel("Data Points")
  plt.ylabel("Binomial(Probability)")
  plt.show()

binomial()
```

```python
import matplotlib.pyplot as plt
def factorial(n):
  fac=1
  for k in range(1,n+1):
    fac=fac*k
  return fac
def poisson():
  p=float(input("Enter the probability(p-->0)::"))
  m=int(input("Enter the number of elements(m-->infinity)"))
  lamda=p*m
  l=[]
  for k in range(m):
    l.append(k)
  poisson=[]
  for k in range(m):
    poisson.append((2.718)**(-1*lamda)*((lamda)**k/factorial(k)))
  plt.bar(l,poisson,color='orange',alpha=0.6,width=0.5)
  plt.title("Probability Mass Function of Poisson Random Variable")
  plt.xlabel("Data Points")
  plt.ylabel("Poisson(Probability)")
  plt.show()

poisson()
```

```python
import matplotlib.pyplot as plt
def geometric():
  n=int(input("Enter the number of elements::"))
  l=[]
  for k in range(1,n+1):
    l.append(k)
  p=float(input("Enter the probability::"))
  geometric=[]
  for k in range(1,n+1):
    geometric.append(((1-p)**(k-1))*p)
  plt.bar(l,geometric,color='pink',width=0.5)
  plt.title("Probability Mass Function of Geometric Random Variable")
  plt.xlabel("Data Points")
  plt.ylabel("Geometric(Probability)")
  plt.show()

geometric()
```
